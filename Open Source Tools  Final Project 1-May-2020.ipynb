### The python code below will generate a random number, use that number as a dictionary key, and print the related value

```python
# create a dictionary

play_dict = {1:"apple", 2:"banana", 3: "carrot", 4: "daikon", 5:"eggplant"}

# generate a random number in the range 1-5

import numpy as np
np.random.seed()
a = np.random.randint(1,5)

# lookup the value associated with the randomly selected number

print(str(a) + " returns " + play_dict[a])
```
