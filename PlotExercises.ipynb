```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import cm
```

```python
data = {"Name":['aa', 'bb', 'cc', 'dd', 'ee'], "Col1": [100, 200, 250, 275, 400], "Col2":['even', 'even', 'even', 'odd', 'even'], "Col3":[1, 2, 4, 99, 43]}
df=pd.DataFrame(data)
df.head
```

```python
df.set_index("Name")
```

```python
df.plot(kind="area", stacked=False)
```

```python
transparency = 0.35 
df.plot(kind='area', alpha=transparency, figsize=(20, 10))

plt.title('Plot Title')
plt.ylabel('Vertical Axis Label')
plt.xlabel('Horizontal Axis Label')

plt.show()
```

```python

df['Col1'].plot(kind="pie", label="label")
```

```python
labels=df['Name']
sizes=df['Col1']
explode = (0, 0.1, 0, 0, 0)
colors = ['salmon', 'gold', 'lightskyblue', 'slategrey', 'brown']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, counterclock = False, colors=colors, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```
